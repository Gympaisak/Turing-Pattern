```python
# Amandas funktion f√∂r gamma
def gamma_func(B, A):
    # from theory we have that the A matrix is n x n, and we go from 0 to n-1 when computing the reachability matrix tau ( in our case gamma)
    n = A.shape[0]
    Bnm = B.shape[:]
    # Creating the reachability matrix gamma w zeros
    gamma_shape = (Bnm[0],(Bnm[0]*Bnm[1]))
    gamma = np.zeros(gamma_shape)
    #print(gamma_shape)
    for c in range(0, n):
        gamma[:,c*B.shape[1]:(c+1)*B.shape[1]] = np.matmul(np.linalg.matrix_power(A, c), B)
    gamma_rank = np.linalg.matrix_rank(gamma, tol=10**(-16))
    print(gamma_rank)
    return gamma_rank, gamma_shape

# tol=10**(-16)
```
